```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import metrics
```

```python
tset = pd.read_csv('train.csv')
tset.head()
```

```python
tset.drop(['Ticket','PassengerId'], 1, inplace=True)
tset.head()
```

```python
tset.info()
```

```python
def assignDeckValue(CabinCode):
    if pd.isnull(CabinCode):
        category='unknown'
    else:
        category=CabinCode[0]
    return category
```

```python
Deck=np.array([assignDeckValue(cabin) for cabin in tset['Cabin'].values])
tset=tset.assign(Deck=Deck)
```

```python
tset['familySize']=tset['SibSp'] + tset['Parch']+1
```

```python
tset.head()
```

```python
tset['Title']=tset.Name.str.extract('([A-Za-z]+)\.', expand=False)
tset['Title']=tset['Title'].replace(['Dr','Rev','Col','Major','Countess','Sir','jonkheer','Lady','Capt','Don'],'Others')
tset['Title']=tset['Title'].replace('Ms','Miss')
tset['Title']=tset['Title'].replace('Mme','Mrs')
tset['Title']=tset['Title'].replace('Mile','Miss')
```

```python
tset.drop(['Cabin','Name'],1,inplace=True)
```

```python
tset.head()
```

```python
tset['Embarked'].value_counts()
common='S'
tset['Embarked']=tset['Embarked'].fillna('S')
tset['Embarked'].isnull().sum()
```

```python
tset.head()
```

```python
means=tset.groupby('Title')['Age'].mean()
title_list=['Master','Miss','Mr','Mrs','Others']
def age_missing_replace(means,dframe,title_list):
    for title in title_list:
        temp=dframe['Title']==title
        dframe.loc[temp,'Age']=dframe.loc[temp,'Age'].fillna(means[title])
age_missing_replace(means, tset, title_list)
```

```python
tset['Embarked']=tset['Embarked'].map({'C':0, 'Q':1, 'S':2})
tset['Sex']=tset['Sex'].map({'male':0, 'female':1})
tset['Title']=tset['Title'].map({'Master':0, 'Miss':1, 'Mr':2, 'Mrs':3, 'Others':4})
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
tset['Deck']=le.fit_transform(tset['Deck'])
```

```python
tset.head()
```

```python
X=tset[['Pclass','Sex','Age','Fare','Embarked','Deck','familySize']].values
Y=tset['Survived'].values
```

```python
svcc=SVC(kernel='linear')
svcc.fit(X,Y)
```

```python
Ypr=svcc.predict(X)
df = pd.DataFrame({'Actual': Y.flatten(), 'Predicted': Y.flatten()})
df
```

```python
testset=pd.read_csv('test.csv')
```

```python
print(metrics.accuracy_score(Y,Ypr))
print(metrics.precision_score(Y,Ypr))
print(metrics.recall_score(Y,Ypr))
```
